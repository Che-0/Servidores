# Resumen de "Integraciones.pdf"

Este PDF de 1 página explica en español la **integración entre servidores web (ej. Apache, Nginx) y servidores de aplicaciones** para generar contenido dinámico. No siempre hay un HTML estático; se necesita ejecutar código (Python, Java, etc.) en tiempo real.

## Rol del Servidor Web y de Aplicaciones
- Servidor web: "Puerta de entrada" para solicitudes; pasa info a servidor de aplicaciones y recibe respuesta.
- Servidor de aplicaciones: Ejecuta código para contenido dinámico (ej. en Python, Java).

## Mecanismos Comunes de Integración
- **CGI (Common Gateway Interface)**: Método más antiguo. Servidor web lanza un nuevo proceso por solicitud. Ineficiente y lento con alto tráfico.
- **FastCGI**: Mejora de CGI. Mantiene procesos persistentes listos para solicitudes. Más rápido y eficiente; usado en PHP.
- **WSGI (Web Server Gateway Interface)**: Estándar para Python. Define interfaz entre servidor web y apps Python; compatible con cualquier framework (ej. Django, Flask).
- **AJP (Apache JServ Protocol)**: Protocolo binario de Apache para comunicación optimizada entre servidor web (Apache HTTP) y servidor Java (ej. Tomcat). Más eficiente que HTTP para tráfico interno.

## Observaciones
- El foco es en eficiencia: Evitar procesos nuevos por solicitud para manejar tráfico alto.
- Incluye diagramas simples (servidor web → cliente; servidor web → app server).
