## Virtual Hosts
Permiten que un único servidor web aloje múltiples sitios web en lugar de requerir un servidor físico por sitio (ej. sitio1.com, sitio2.org y blog.sitio1.com en el mismo servidor).

**Tipos principales**:
- **Basado en nombre** (más común): El servidor distingue sitios por el dominio solicitado (lee la cabecera "Host" en la solicitud HTTP). Ej. Cuando el navegador pide www.sitio1.com, el servidor sabe qué contenido servir.
- **Basado en IP**: Cada sitio tiene su propia IP. El servidor usa la IP de destino para decidir. Menos común hoy debido a la escasez de IPv4.

## Proxy Server (Proxy Directo o de Reenvío)
Actúa como intermediario para la seguridad del cliente. Cuando un cliente (máquina local) solicita algo de internet, el proxy envía la solicitud en su nombre y fetches los datos.

**Casos de uso**:
- **Caching**: Almacena contenido para servirlo rápido sin ir a internet cada vez, ahorrando ancho de banda.
- **Anonimato**: El servidor final no sabe el origen real de la solicitud.
- **Seguridad**: Actúa como firewall; filtra o bloquea solicitudes malas. Múltiples clientes se conectan a un proxy centralizado.
- **Encriptación y Desencriptación**: Enmascara IP y datos sensibles.
- **Logging**: Registra páginas visitadas para optimizar caching.
- **Microservicios**: Proxies como sidecar, HTTP o TCP se despliegan junto a aplicaciones para manejar networking (ej. en contenedores).

Se usa en redes internas: El cliente se configura para que todas las peticiones salgan por el proxy. Útil para filtrado de contenido, caching y seguridad.

## Reverse Proxy Server (Proxy Inverso)
Es lo opuesto al proxy directo: Protege al servidor en lugar del cliente. Las solicitudes de internet van al reverse proxy, que decide a qué servidor interno redirigirlas (sin exponer servidores reales).

**Ventajas (además de caching y seguridad del proxy normal)**:
- **Load Balancing**: Distribuye carga entre servidores; comprime (zipping) solicitudes para mejorar rendimiento.
- **Ingress**: En Kubernetes/microservicios, enruta solicitudes a servicios específicos (ej. si es para "Picture API", va al servicio correspondiente). Actúa como router.
- **Canary Deployment**: Enruta un porcentaje de tráfico (ej. 10%) a versiones experimentales (ej. thumbnails random).

Se coloca delante de varios servidores web. Los clientes de internet se conectan al proxy, que distribuye a internos.

**Diferencia con Load Balancer**: Un reverse proxy puede ser load balancer, pero no al revés. El load balancer solo balancea carga; el reverse proxy funciona con un solo servidor y añade caching, seguridad, etc.

**¿Se pueden usar proxy y reverse proxy juntos?**: Sí, en "service mesh" un proxy actúa como ambos.

## Application Layer Gateways (ALG)
Tipo avanzado de proxy. Mientras un proxy normal maneja protocolos básicos (HTTP, etc.), un ALG inspecciona y modifica tráfico de protocolos complejos (ej. FTP, SIP para VoIP). Asegura que la comunicación funcione ajustando información de red.

**Relación con Proxies**:
- Todos los ALGs son proxies (intermediarios).
- Diferencia: ALG tiene "inteligencia" profunda en el protocolo; no solo reenvía/filtra, sino que manipula paquetes para conectividad correcta.
