# Resumen del PDF: "Http.pdf"

Este PDF es un documento de 2 páginas que explica los fundamentos del protocolo HTTP, incluyendo su funcionamiento cliente-servidor, métodos comunes, códigos de estado y la evolución de sus versiones (HTTP/1.1, HTTP/2 y HTTP/3). Parece ser notas educativas o apuntes, con algunos errores tipográficos en los códigos de estado (por ejemplo, "300 OK" en lugar de "200 OK"). A continuación, un resumen detallado estructurado por secciones principales.

## Funcionamiento Básico de HTTP
- **Modelo Cliente-Servidor**: HTTP opera en modo cliente-servidor. Un cliente (ej. navegador web) envía una solicitud a un servidor (donde se aloja una página web), y este responde.
- **Elementos Clave**:
  - **Métodos HTTP**: Indican la acción que se realiza.
    - **GET**: Solicita datos o recursos específicos. Es seguro e idempotente (no modifica datos y se puede repetir sin problemas).
    - **POST**: Envía datos al servidor, pudiendo modificar su estado (a diferencia de GET).
  - **Códigos de Estado**: Son respuestas de 3 dígitos (el PDF menciona erróneamente 5 dígitos). El primer dígito indica el tipo:
    - **1xx (Informativa)**: Solicitud recibida y en proceso.
    - **2xx (Éxito)**: Solicitud recibida, entendida y aceptada.
    - **3xx (Redirección)**: Requiere acción adicional (ej. recurso movido).
    - **4xx (Error del Cliente)**: Solicitud con error (ej. 404 Not Found).
    - **5xx (Error del Servidor)**: Solicitud válida, pero servidor no puede completarla.

- **Ejemplos de Interacción**:
  - Al escribir una URL, el navegador envía un GET.
  - Posibles respuestas del servidor:
    - 200 OK: Recurso entregado.
    - 301 Moved Permanently: Recurso movido a nueva dirección.
    - 404 Not Found: Recurso no existe.

## Códigos de Estado Más Comunes
- **200 OK**: Solicitud exitosa (el PDF dice "300 OK", probable error).
- **201 Created**: Solicitud exitosa y nuevo recurso creado (común en POST; PDF dice "301 Created", error).
- **301 Moved Permanently**: Recurso asignado a nueva URL permanente.
- **302 Found**: Recurso temporalmente en otra ubicación (mantener URL original).
- **400 Bad Request**: Sintaxis incorrecta en la solicitud.
- **401 Unauthorized**: Requiere autenticación.
- **404 Not Found**: Recurso no encontrado.
- **500 Internal Server Error**: Error genérico en el servidor.
- **503 Service Unavailable**: Servidor no disponible (sobrecarga o mantenimiento).

## Versiones de HTTP
### HTTP/1.1 (El Clásico)
- Evolución de HTTP/1.0.
- Permite enviar el nombre de la página en cabeceras para que el servidor responda según la URL.
- Modelo: Abre conexión TCP, envía solicitud, espera respuesta y cierra (en la mayoría de casos).
- **Problemas**: Cada recurso (imágenes, CSS, JS) requiere una solicitud separada, lo que ralentiza páginas modernas con muchos elementos. En páginas simples antiguas, no era un issue.

### HTTP/2 (La Evolución)
- Soluciona limitaciones de HTTP/1.1.
- **Características Principales**:
  - **Multiplexación**: Envía múltiples solicitudes y respuestas por una sola conexión, evitando congestión.
  - **Compresión de Cabeceras**: Reduce tamaño de cabeceras repetitivas.
  - **Server Push**: Servidor envía recursos anticipadamente (ej. CSS/JS). En práctica, frágil; reemplazado por "Link: rel=preload" y "103 Early Hint" debido a problemas con caché y proxies.

### HTTP/3 (La Revolución)
- Versión más reciente, con cambio fundamental.
- **Protocolo QUIC**: Basado en UDP (en lugar de TCP de versiones anteriores).
- **Ventajas**: Si un paquete se pierde, solo afecta ese stream de datos, no toda la conexión. Ideal para conexiones móviles o inestables (TCP detiene todo si hay pérdida).
