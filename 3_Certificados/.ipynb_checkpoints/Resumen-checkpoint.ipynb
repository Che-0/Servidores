
### 1. X509_260110_050130.pdf – Certificado Digital X.509
- **Definición**: Un certificado digital es una credencial electrónica que autentica la identidad de una persona, organización o entidad en internet.  
- **Contenido principal**:
  - Nombre de la entidad (sujeto).
  - Clave pública.
  - Firma digital de una autoridad confiable (CA).
- **Usos**: Cifrado de datos, autenticación de sitios web (HTTPS), firmas electrónicas. Garantiza seguridad y confianza en comunicaciones/transacciones online.
- **Estándar**: X.509 (el más usado para certificados digitales).
- **Campos clave**:
  - **Subject (Sujeto)**: Identifica al titular (nombre, organización, ubicación).
  - **Issuer (Emisor)**: La CA que lo emitió y firmó.
  - **Validity (Validez)**: Periodo de tiempo en que el certificado es válido (fecha inicio-fin).
- **Observación**: El PDF menciona que se usa para "garantizar la seguridad y la confianza", lo cual es clave en TLS/SSL.

### 2. Csr_260110_043031.pdf – Certificate Signing Request (CSR)
- **Definición**: En PKI, una CSR (Certificate Signing Request) es un mensaje que envía un solicitante (tú o tu servidor) a una CA para pedir un certificado digital de identidad.
- **Proceso paso a paso** (exacto como lo describes):
  1. Genera un par de claves: **Clave Privada** (se queda contigo, nunca se comparte) y **Clave Pública**.
  2. Crea la CSR usando la clave pública y proporcionando info requerida (ej. nombre de dominio, organización, país).
  3. Envía la CSR a una **Autoridad de Certificación (CA)** para que la firme y genere el certificado completo.
- **Nota**: La CA verifica la identidad y firma la CSR con su clave privada. El certificado resultante incluye tu clave pública + firma de la CA.

### 3. Ca_260110_041954.pdf – Certificate Authority (CA)
- **Contenido**: La página está en blanco (o vacía). Probablemente sea un PDF incompleto o solo título.  
  - Si es intencional, podría ser un placeholder para notas sobre CA.  
  - En tus notas manuscritas (imagen del CA) expandes mucho:  
    - **CA (Certificate Authority)**: Entidad que emite y firma certificados.  
    - **Tipos**: Root CA (raíz, nivel superior) e Intermediate CA (subordinadas, crean cadena de confianza).  
    - **Ejemplos**: Let's Encrypt, GeoTrust, DigiCert, Symantec.  
    - **Funciones**: Emisión (verifica identidad), Firma (con su clave privada), Revocación (si hay compromiso).  
    - **Cadena**: Root CA → Intermediate CA → Certificado final (usuario).

## Resumen General de Temas Relacionados (de tus notas manuscritas)
Tus notas complementan perfecto estos PDFs y cierran el ciclo de PKI:
- **CSR** → Generas claves + CSR → Envías a CA → CA emite certificado X.509 firmado.
- **X.509** → Estructura del certificado (Subject, Issuer, Validity, etc.).
- **CA** → Jerarquía (Root → Intermediate), funciones (emisión, firma, revocación).
- Todo encaja en **PKI** (Public Key Infrastructure): Confidencialidad, firmas digitales, verificación/revocación, gestión de claves.
